```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os,sys
import scipy.stats as stats
import seaborn as sns


immigration_data = 'abc.csv'
df = pd.read_csv(immigration_data)
df.head()
```

```python
df = df.rename(columns={"Region/Sub-Region or Country of Birth": "Country"})
df.head()
```

```python
df.set_index('Country', inplace=True)
df.head()
```

```python
# Clean data
df = df.fillna(0)
```

```python
df['1999'].value_counts()
```

```python
df = df.replace('D', '0', regex=True)
```

```python
df = df.replace('X', '0', regex=True)
```

```python
df[df.columns] = df[df.columns].apply(pd.to_numeric, errors='coerce')
```
